## Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

## Importing Train Data

```python
train=pd.read_csv('train.csv')
train.head()
```

```python
train.isnull().sum()
```

```python
train.count()
'''the null value rows can be removed because they are both categorical values and 
since they contribute to a very small fraction,i.e. education =2409/52399 & Previous_year_rating=4124/50684'''
```

```python
trdata=pd.DataFrame(train)
```

```python
trdata.dropna(inplace=True)
```

```python
trdata.isnull().sum()
```

```python
trdata.head(100)
```

## Converting Categorical Values to Dummy Values 

```python
Dept=pd.get_dummies(trdata["department"],drop_first=True)
Dept.head()
Reg=pd.get_dummies(trdata["region"],drop_first=True)
Reg.head()
Edu=pd.get_dummies(trdata["education"],drop_first=True)
Edu.head()
Sex=pd.get_dummies(trdata["gender"],drop_first=True)
Sex.head()
RC=pd.get_dummies(trdata["recruitment_channel"],drop_first=True)
RC.head()
KPI80=pd.get_dummies(trdata["KPIs_met >80%"],drop_first=True)
KPI80.head()
AW=pd.get_dummies(trdata["awards_won?"],drop_first=True)
AW.head()
```

## Removing Old Categorical Values

```python
#trdata.drop(['department','region','education','gender','recruitment_channel','KPIs_met >80%','awards_won?'],axis=1,inplace=True)
trdata.head()
```

## Adding New Dummy Value columns

```python
trdata=pd.concat([trdata,Dept,Reg,Edu,Sex,RC,KPI80,AW],axis=1)
```

```python
trdata.head()
```

## Removing employee_id column to build the Logistic Regression Model

```python
trdata.drop("employee_id",axis=1,inplace=True)
trdata.head()
```

## Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
```

```python
x=trdata.drop('is_promoted',axis=1)
y=trdata['is_promoted']
```

```python
logmodel.fit(x,y)
```
